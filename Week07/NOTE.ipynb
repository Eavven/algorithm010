Q:高级搜索相比其他搜索高级在何处？

A:朴素搜索类似深度优先或者广度优先，会构建一颗完整的树，但是有些分支里并没有我们需要的答案，又或者这些分支在此前已经生成过一次，如果要再生成等同于做重复性工作。

高级搜索在此进行优化，一个方向是去除不必要工作，即对于已知不可能会有答案的分支直接放弃，即剪枝，以及记录之前已经产生的分支信息，以防后续又产生同批一样的，避免重复性工作。

另一个方向是增加搜索方向，如双向搜索以及A星搜索

Q:有了二叉查找树、平衡树（AVL）为啥还需要红黑树？
    
A:平衡树要求过高，稍有不平衡就要调整，维护成本较高。而对于一些要求不严格的情况，只需要做到大致平衡时，红黑树就可以登场了。它是一种近似平衡的二叉搜索树，能够确保任何一个节点的左右子树高度差小于两倍。

Q:红黑树的性质？其实际应用场景
A:根节点和叶子节点是黑色，两个红色节点不能相互邻接，从任一节点到其每个叶子的所有路径都包含相同数目的黑色节点。
